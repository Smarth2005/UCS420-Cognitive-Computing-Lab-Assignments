**(g) Add a new column that categorizes employees based on their performance rating:**   
**i. Excellent for ratings >= 4.5**   
**ii. Good for ratings >= 4.0 but < 4.5**  
**iii. Average for ratings < 4.0** 


```python
# Categorize based on Rating
df['Category'] = df['Rating'].apply(lambda x: 'Excellent' if x >= 4.5 else 'Good' if x >= 4.0 else 'Average')
```

```python
df
```

```python
df.drop(columns = ['Category'], inplace = True)
```

```python
df
```

```python
# Alternatively, categorization can be done using numpy.select()
import numpy as np
conditions = [
    df['Rating'] >= 4.5,
    (df['Rating'] >= 4.0) & (df['Rating'] < 4.5),
    df['Rating'] < 4.0
]

choices = ['Excellent', 'Good', 'Average']

df['Category'] = np.select(conditions, choices)

df
```
