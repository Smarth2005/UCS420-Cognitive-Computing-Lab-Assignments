#### ðŸ“Œ **What is Correlation in EDA?**
  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Correlation helps you:
- Understand how strongly two variables are related.
- Spot variables that move together (positively or negatively).
- Identify redundant features (those that give similar info).

#### **ðŸ“Œ Why is Correlation useful in EDA?**
- **Feature selection:**
You can drop or combine highly correlated features to simplify models (e.g., donâ€™t keep both if Votes and Ratings are almost the same).

- **Multicollinearity check:**
Before regression or ML, avoid using predictors that are strongly correlated â€” it messes up coefficients!

- **Data understanding:**
You see which factors influence each other (e.g., is Average Cost related to Rating?).

- **Heatmaps make it visual:**
You can spot strong/weak relationships easily with colors.



```python
df.corr()
```

```python
df_numeric = df.select_dtypes(include=['number'])  # Select only numeric columns
corr_matrix = df_numeric.corr()
print(corr_matrix)
```

```python
sns.heatmap(corr_matrix)
```

### **Joint Plot**
A joint plot allows to study the relationship between 2 numeric variables. The central chart displays their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

### **Pair Plot**

A "pair plot" is also known as a scatter plot, in which one variable in the same data row is matched with another variable's value, like this: Plair Plots are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

```python
sns.pairplot(df,hue = 'smoker')
```
