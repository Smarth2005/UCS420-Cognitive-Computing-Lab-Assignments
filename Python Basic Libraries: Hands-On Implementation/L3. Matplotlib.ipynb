## Visualization of Data and Model Fit using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple examples
x = np.arange(0,10)
y = np.arange(11,21)
```

### **Creating Subplots**

```python
x = np.arange(0,10)
y = np.square(x)

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.title('Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.title('Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2,2,3)
plt.plot(x,y,'bo', markersize = 4)
plt.title('Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2,2,4)
plt.plot(x,y,'m*')
plt.title('Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
fig, ax = plt.subplots(1,2) # nrows, ncols
ax[0].plot(x,y,'c')
ax[0].set_title('Plot 1')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')

ax[1].plot(x,y,'purple')
ax[1].set_title('Plot 2')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'maroon',label = 'Sine waveform')
plt.plot(x,z,'orange',label = 'Cosine waveform')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('$y = sin(x), z = cos(x)$')

plt.grid(True, linewidth = 0.5)
plt.axhline(y = 0, color = 'black', linewidth = 1)
plt.axvline(x = 0, color = 'black', linewidth = 1)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Custom Data Array
data = np.array([25, 98, 63, 14, 36, 87, 11, 20, 55, 5, 79, 64, 55, 10, 31, 27])

# Define Number of Bins (You can adjust it)
num_bins = 8  # Choose a suitable number of bins

# Plot Histogram & Get Bin Information
counts, bin_edges, _ = plt.hist(data, bins=num_bins, color='coral', edgecolor='black', alpha=0.8, label="Histogram")

# Compute Bin Centers (Midpoints of Each Bar)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Plot Frequency Polygon (Connecting Histogram Bar Centers)
plt.plot(bin_centers, counts, color='purple', marker='*', linestyle='--', linewidth=2, markersize=8, label="Frequency Polygon")

# Labels & Title
plt.title("Histogram with Frequency Polygon")
plt.xlabel('Data Range')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
data = np.array([25,98,63,14,36,87,11,20,55,5,79,64,55,10,31,27])

num_bins = int(1 + np.log2(len(data)))  # Sturges' rule
plt.hist(data, bins = num_bins, color = 'lightgreen', edgecolor ='black', label = 'Histogram')

counts, bin_edges = np.histogram(data, bins = num_bins)

bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
plt.plot(bin_centers, counts, color = 'red', marker = 'o', linestyle = '--', label = 'Frequency plot')

plt.title('Histogram with Frequency plot')
plt.xlabel('Data Range')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

### **Pie-Chart**

```python
# Data to plot
labels = ['Python', 'C++', 'Java', 'Ruby']
sizes  = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0)

# Plot
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)

plt.axis('equal')
plt.show()
```
