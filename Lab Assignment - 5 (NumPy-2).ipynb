**Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), find**  
**i. Sum of all elements**   
**ii. Sum of all elements row-wise**   
**iii. Sum of all elements column-wise**   

```python
import numpy as np
```

```python
mat = np.matrix('[4,1,9;12,3,1;4,5,6]')
print("Matrix =")
print(mat)
print()

s = mat.sum()
row_sum = mat.sum(axis = 1)
col_sum = mat.sum(axis = 0)

print("Sum of all elements =",s)
print("Sum of all elements row-wise =")
print(row_sum)
print("Sum of all elements column-wise =",col_sum)
```

**Q.2(a) For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find**   
**i. Sorted array**   
**ii. Indices of sorted array**   
**iii. 4 smallest elements**   
**iv. 5 largest elements** 
  
**(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find**   
**i. Integer elements only**   
**ii. Float elements only**   

```python
arr1 = np.array([10,52,62,16,16,54,453])
arr1_sorted = np.sort(arr1)
sorted_indices = np.argsort(arr1)
smallest_elements = arr1[:4]
largest_elements = arr1[-5:]

print("(a)")
print("Original Array          :",arr1)
print("Sorted Array            :",arr1_sorted)
print("Indices of sorted array :",sorted_indices)
print("4 Smallest elements     :",smallest_elements)
print("5 largest elements      :",largest_elements)
print()

arr2 = np.array([1.0,1.2,2.2,2.0,3.0,2.0])

# Finding integer and float elements
# Approach-1

# An element is an integer if its fractional part is zero and a float if its fractional part is non-zero.
# We'll use np.modf(), which splits an array into fractional and integer parts:
fractional, _ = np.modf(arr2)
integer_elements = arr2[fractional == 0]
float_elements = arr2[fractional != 0]

print("(b)")
print("Given Array     :",arr2)
print("Integer Elements:", integer_elements)
print("Float Elements  :", float_elements)

# Approach-2 [Alternative Approach using astype(int)
int_ele = arr2[arr2 == arr2.astype(int)]
float_ele = arr2[arr2 != arr2.astype(int)]

print("By Alternative Approach using astype(int)")
print("Integer Elements:", int_ele)
print("Float Elements  :", float_ele)
```

**Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting.**   
**a) Generate your unique sales dataset:**   
- **Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65 + 66 = 131 → sales = [131, 181, 231, 281, 331].)**  
- **Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200].**

**b) Compute your personalized tax rate as ((X % 5) + 5) / 100. Use broadcasting to apply this tax rate to each sales value.**

**c) Adjust sales based on discount:** 
- **If sales < X+100, apply a 5% discount.**
- **If sales >= X+100, apply a 10% discount.**

**d) Expand sales data for multiple weeks:**
- **Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.**
- **Increase sales by 2% per week using element-wise broadcasting.**

```python
# Compute X (sum of ASCII values of initials)
X = ord('S') + ord('K') 

# Generate sales dataset
sales = np.array([X, X + 50, X + 100, X + 150, X + 200])
print("Sales Dataset:", sales)
```

```python
# Compute tax rate
tax_rate = ((X % 5) + 5) / 100

# Apply tax rate
tax_amount = sales * tax_rate
sales_with_tax = sales + tax_amount

print("Tax Rate                 :", tax_rate)
print("Tax Amount on Sales      :", tax_amount)
print("Sales after applying tax :", sales_with_tax)
```

```python
# Create a copy of sales to apply discount
discounted_sales = sales.copy()

# Apply discounts
discounted_sales[sales < X + 100] *= 0.95   # 5% discount
discounted_sales[sales >= X + 100] *= 0.90  # 10% discount

print("Discounted Sales:", discounted_sales)
```

##### Tip: When you multiply an integer array by 0.95 or 0.90, NumPy tries to store a floating-point result into an integer array, which causes an error (ufunc type error) because NumPy arrays with integer data types do not support in-place floating-point operations.

```python
# Create a copy of sales to apply discount
discounted_sales = sales.astype(float)

# Apply discounts
discounted_sales[sales < X + 100] *= 0.95   # 5% discount
discounted_sales[sales >= X + 100] *= 0.90  # 10% discount

print("Discounted Sales:", discounted_sales)
```

```python
# Stack sales for 3 weeks
weekly_sales = np.vstack([sales] * 3)  # 3 rows, 5 columns

print("Weekly Sales Data:\n", weekly_sales)
```

```python
# Create a scaling factor for each week
growth_factors = np.array([1.00, 1.02, 1.04]).reshape(3, 1)

# Apply the increase
weekly_sales = weekly_sales * growth_factors

print("Updated Weekly Sales:\n", weekly_sales)
```

**Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy:** 
- **Y = x<sup>2</sup>**
- **Y = sin(x)** 
- **Y = e<sup>x</sup>**
- **Y = log(|x| + 1)**

**Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -10 to 10 with 100 points
x = np.linspace(-10, 10, 100)

# Compute y values for each function
y1 = x**2                  # Y = x^2
y2 = np.sin(x)             # Y = sin(x)
y3 = np.exp(x)             # Y = e^x
y4 = np.log(np.abs(x) + 1) # Y = log(|x| + 1)

# List of functions and corresponding titles
functions = [y1, y2, y3, y4]
titles = [r"$Y = x^2$", r"$Y = \sin(x)$", r"$Y = e^x$", r"$Y = \log(|x| + 1)$"]
colors = ["b", "r", "g", "m"]

# Create a figure with subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Loop through each axis and corresponding function
for i, ax in enumerate(axes.flat):
    ax.plot(x, functions[i], color=colors[i])    # Plot the function
    ax.set_title(titles[i])                      # Set the title
    ax.set_xlabel("x values")                    # Set the x-label
    ax.set_ylabel("y values")                    # Set the y-label
    ax.grid(True)                                # Add grid
    ax.axhline(0, color='black', linewidth=0.5)  # Horizontal axis line
    ax.axvline(0, color='black', linewidth=0.5)  # Vertical axis line

# Adjust layout for spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
